```python
import requests
import pandas as pd
import datetime
```

# fetch data

```python
parameters = {"page": 1, "perPage": 10000}
response_matches = requests.get("https://api.godsunchained.com/v0/match", params=parameters)
df_total = pd.DataFrame(response_matches.json()['records'])
df_total
```

# feature engineering

```python
df= pd.read_csv("matches.csv")
df = df.drop("Unnamed: 0", axis= 1)
df
```

```python
def unix2iso(unix):
    return datetime.datetime.fromtimestamp(unix).strftime('%Y-%m-%d %H:%M:%S')
df['start_time'] = df['start_time'].apply(unix2iso)
df['end_time'] = df['end_time'].apply(unix2iso)
```

```python
df_unique = df.drop_duplicates()
df_unique = df_unique[df_unique['game_mode']==13]
df_unique
```

```python
df_unique.to_csv("df_unique.csv")
```
