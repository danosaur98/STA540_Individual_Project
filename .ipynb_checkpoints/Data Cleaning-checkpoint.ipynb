# feature engineering


```python
df_unique = pd.read_csv("df_unique.csv")
df_unique = df_unique.drop("Unnamed: 0", axis= 1)
df_unique = df_unique.sort_values(by="start_time")
df_unique = df_unique.reset_index(drop=True)
df_unique.head()
```

```python
df = df_unique.head(1000)
df.head()
```

# all unique card IDs

```python
all_unique_cards = set()
for i,row in df_unique.iterrows():
    player_info = eval(row['player_info'])
    all_unique_cards = all_unique_cards.union(set(player_info[0]['cards']))
    all_unique_cards = all_unique_cards.union(set(player_info[1]['cards']))
```

# card info

```python
cards = pd.read_csv("cards.csv")
cards = cards.drop("Unnamed: 0", axis=1)
cards.head()
```

```python
cards_map = {}
for i, row in cards.iterrows():
    cards_map[row['id']] = row
```

```python
player_id2skill = {}

def get_player_skill(player_id):
    '''
    create empty template given a player id
    '''
    if player_id not in player_id2skill:
        player_id2skill[player_id] = {
            'num_games_played': 0,
            'num_games_played_per_god': {
                'nature': 0,
                'death': 0,
                'deception': 0,
                'light': 0,
                'magic': 0,
                'war': 0
            },
            'num_games_won': 0,
            'num_games_won_per_god': {
                'nature': 0,
                'death': 0,
                'deception': 0,
                'light': 0,
                'magic': 0,
                'war': 0
            }
        }
    return player_id2skill[player_id]

def update_player_skill(player_id, did_win, god):
    player_id2skill[player_id]['num_games_played'] += 1
    player_id2skill[player_id]['num_games_played_per_god'][god] += 1
    if did_win:
        player_id2skill[player_id]['num_games_won'] += 1
        player_id2skill[player_id]['num_games_won_per_god'][god] += 1
    
    
```

```python
def construct_player_card_info(player_info):
    '''

    given player info, construct the following columns:
   'player_god',
   'player_god_power',
   'player_deck_avg_mana_cost',
   'player_deck_num_creatures',
   'player_deck_num_spells',
   'player_deck_num_weapons',
   'player_deck_avg_creature_strength',
   'player_deck_avg_creature_health', 
   'player_card_i',
    '''
    player_god = player_info['god']
    player_god_power = player_info['god_power']
    player_deck_avg_mana_cost = 0
    player_deck_num_creatures = 0
    player_deck_num_spells = 0
    player_deck_num_weapons = 0
    player_deck_avg_creature_strength = 0
    player_deck_avg_creature_health = 0
    player_card_k_dict = {}
    
    total_mana_cost = 0
    total_creature_strength = 0
    total_creature_health = 0
    
    # create empty dictionary of all unique cards
    for uid in all_unique_cards:
        player_card_k_dict[uid] = 0
    
    # iterate through card id's in player's deck
    for uid in player_info['cards']:
        card_info = cards_map[uid]
        player_card_k_dict[uid] += 1
        total_mana_cost += card_info['mana']
        if card_info['type'] == "creature":
            player_deck_num_creatures += 1
            if eval(card_info['attack'])['Valid']:
                total_creature_strength +=  eval(card_info['attack'])['Int64']
            if eval(card_info['health'])['Valid']:
                total_creature_health +=  eval(card_info['health'])['Int64']
        elif card_info['type'] == 'spell':
            player_deck_num_spells += 1
        elif card_info['type'] == "weapon":
            player_deck_num_weapons += 1\
    
    player_deck_avg_mana_cost = total_mana_cost / 30
    player_deck_avg_creature_strength = total_creature_strength / player_deck_num_creatures
    player_deck_avg_creature_health = total_creature_health / player_deck_num_creatures

    
    return {"player_god": player_god,
            "player_god_power": player_god_power,
            "player_deck_avg_mana_cost": player_deck_avg_mana_cost,
            "player_deck_num_creatures": player_deck_num_creatures,
            "player_deck_num_spells": player_deck_num_spells,
            "player_deck_num_weapons": player_deck_num_weapons,
            "player_deck_avg_creature_strength": player_deck_avg_creature_strength,
            "player_deck_avg_creature_health": player_deck_avg_creature_health,
            "player_card_k_dict": player_card_k_dict}
```

```python
colnames = ['outcome', #1 if player i wins over player j
           'start_time',
           'player_i_id',
           'player_i_god',
           'player_i_god_power',
           'player_i_god_win_ratio',
           'player_i_god_games_played',
           'player_i_games_played',
           'player_i_deck_avg_mana_cost',
           'player_i_deck_num_creatures',
           'player_i_deck_num_spells',
           'player_i_deck_num_weapons',
           'player_i_deck_avg_creature_strength',
           'player_i_deck_avg_creature_health',           
           'player_j_id',
           'player_j_god',
           'player_j_god_power',
           'player_j_god_win_ratio',
           'player_j_god_games_played',
           'player_j_games_played',
           'player_j_deck_avg_mana_cost',
           'player_j_deck_num_creatures',
           'player_j_deck_num_spells',
           'player_j_deck_num_weapons',
           'player_j_deck_avg_creature_strength',
           'player_j_deck_avg_creature_health'
           ]
print(len(colnames))
for uid in all_unique_cards:
    colnames.append("player_i_card_{}".format(uid))
for uid in all_unique_cards:
    colnames.append("player_j_card_{}".format(uid))
```

```python
import time
now = time.time()
df_processed = None
player_id2skill = {}
for i, row in df_unique.iterrows():
    print(i, time.time() -now)
    curr_row_dict = {}
    
    player_info = eval(row['player_info'])
    
    #randomly place winner as player i or player j
    if random.random() <= 1:
        player_i = player_info[0]
        player_j = player_info[1]
        curr_row_dict['outcome'] = 1 
    else:
        player_i = player_info[1]
        player_j = player_info[0]
        curr_row_dict['outcome'] = 0

    curr_row_dict['start_time'] = row['start_time']
    player_i_card_info = construct_player_card_info(player_i)
    player_i_skill = get_player_skill(player_i["user_id"])
    
    curr_row_dict['player_i_id'] = player_i["user_id"]
    curr_row_dict['player_i_god'] = player_i_card_info["player_god"]
    curr_row_dict['player_i_god_power'] = player_i_card_info["player_god_power"]
    
    num_games_won = player_i_skill
    curr_row_dict['player_i_god_games_played'] = player_i_skill['num_games_played_per_god'][player_i_card_info["player_god"]]
    curr_row_dict['player_i_god_win_ratio'] = player_i_skill['num_games_won_per_god'][player_i_card_info["player_god"]] / curr_row_dict['player_i_god_games_played'] if curr_row_dict['player_i_god_games_played'] > 0 else 0
    curr_row_dict['player_i_games_played'] = player_i_skill['num_games_played']
    curr_row_dict['player_i_deck_avg_mana_cost'] = player_i_card_info["player_deck_avg_mana_cost"]
    curr_row_dict['player_i_deck_num_creatures'] = player_i_card_info["player_deck_num_creatures"]
    curr_row_dict['player_i_deck_num_spells'] = player_i_card_info["player_deck_num_spells"]
    curr_row_dict['player_i_deck_num_weapons'] = player_i_card_info["player_deck_num_weapons"]
    curr_row_dict['player_i_deck_avg_creature_strength'] = player_i_card_info["player_deck_avg_creature_strength"]
    curr_row_dict['player_i_deck_avg_creature_health'] = player_i_card_info["player_deck_avg_creature_health"]

    update_player_skill(player_id=player_i["user_id"],
                       did_win=curr_row_dict['outcome'],
                       god=player_i_card_info["player_god"])
    
    for card_uid in player_i_card_info["player_card_k_dict"]:
        curr_row_dict["player_i_card_{}".format(card_uid)] = player_i_card_info["player_card_k_dict"][card_uid]
    
    player_i_card_info = construct_player_card_info(player_i)
    player_i_skill = get_player_skill(player_i["user_id"])
    
    curr_row_dict['player_i_id'] = player_i["user_id"]
    curr_row_dict['player_i_god'] = player_i_card_info["player_god"]
    curr_row_dict['player_i_god_power'] = player_i_card_info["player_god_power"]
    
    num_games_won = player_i_skill
    curr_row_dict['player_i_god_games_played'] = player_i_skill['num_games_played_per_god'][player_i_card_info["player_god"]]
    curr_row_dict['player_i_god_win_ratio'] = player_i_skill['num_games_won_per_god'][player_i_card_info["player_god"]] / curr_row_dict['player_i_god_games_played'] if curr_row_dict['player_i_god_games_played'] > 0 else 0
    curr_row_dict['player_i_games_played'] = player_i_skill['num_games_played']
    curr_row_dict['player_i_deck_avg_mana_cost'] = player_i_card_info["player_deck_avg_mana_cost"]
    curr_row_dict['player_i_deck_num_creatures'] = player_i_card_info["player_deck_num_creatures"]
    curr_row_dict['player_i_deck_num_spells'] = player_i_card_info["player_deck_num_spells"]
    curr_row_dict['player_i_deck_num_weapons'] = player_i_card_info["player_deck_num_weapons"]
    curr_row_dict['player_i_deck_avg_creature_strength'] = player_i_card_info["player_deck_avg_creature_strength"]
    curr_row_dict['player_i_deck_avg_creature_health'] = player_i_card_info["player_deck_avg_creature_health"]

    update_player_skill(player_id=player_i["user_id"],
                       did_win=curr_row_dict['outcome'],
                       god=player_i_card_info["player_god"])
    
    for card_uid in player_i_card_info["player_card_k_dict"]:
        curr_row_dict["player_i_card_{}".format(card_uid)] = player_i_card_info["player_card_k_dict"][card_uid]
        player_j_card_info = construct_player_card_info(player_j)
        
    player_j_skill = get_player_skill(player_j["user_id"])
    
    curr_row_dict['player_j_id'] = player_j["user_id"]
    curr_row_dict['player_j_god'] = player_j_card_info["player_god"]
    curr_row_dict['player_j_god_power'] = player_j_card_info["player_god_power"]
    
    num_games_won = player_j_skill
    curr_row_dict['player_j_god_games_played'] = player_j_skill['num_games_played_per_god'][player_j_card_info["player_god"]]
    curr_row_dict['player_j_god_win_ratio'] = player_j_skill['num_games_won_per_god'][player_j_card_info["player_god"]] / curr_row_dict['player_j_god_games_played'] if curr_row_dict['player_j_god_games_played'] > 0 else 0
    curr_row_dict['player_j_games_played'] = player_j_skill['num_games_played']
    curr_row_dict['player_j_deck_avg_mana_cost'] = player_j_card_info["player_deck_avg_mana_cost"]
    curr_row_dict['player_j_deck_num_creatures'] = player_j_card_info["player_deck_num_creatures"]
    curr_row_dict['player_j_deck_num_spells'] = player_j_card_info["player_deck_num_spells"]
    curr_row_dict['player_j_deck_num_weapons'] = player_j_card_info["player_deck_num_weapons"]
    curr_row_dict['player_j_deck_avg_creature_strength'] = player_j_card_info["player_deck_avg_creature_strength"]
    curr_row_dict['player_j_deck_avg_creature_health'] = player_j_card_info["player_deck_avg_creature_health"]

    update_player_skill(player_id=player_j["user_id"],
                       did_win=curr_row_dict['outcome'],
                       god=player_j_card_info["player_god"])
    
    for card_uid in player_j_card_info["player_card_k_dict"]:
        curr_row_dict["player_j_card_{}".format(card_uid)] = player_j_card_info["player_card_k_dict"][card_uid]
    
    curr_row_df = pd.DataFrame.from_dict([curr_row_dict])
    curr_row_df = curr_row_df[colnames]
    
    if df_processed is None:
        df_processed = curr_row_df
    else:
        df_processed = df_processed.append(curr_row_df)
```
